```python
x = np.linspace(0,10,50)
y = 2.5 * x + np.random.normal(size = x.size)
```

```python
A = np.vstack([x, np.ones(x.size)]).T
A
```

```python
#Obtengo la pendiente y la constante
m, c = np.linalg.lstsq(A,y, rcond=None)[0]
```

```python
pendiente = m*x + c
```

```python
#Grafico la regresion lineal
plt.plot(x, pendiente, "r",label = 'Regresion lineal')
plt.scatter(x,y=y, label = 'Datos experimentales')
plt.legend()
plt.show()
```
